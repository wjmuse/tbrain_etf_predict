```python
%matplotlib inline

import os
import numpy as np
import pandas as pd
from collections import OrderedDict
import datetime
import time
from IPython.display import HTML
import matplotlib.pyplot as plt
from etf_tools import kd_rsv, ez_plot, candle_stick, rsi, ema, macd

plt.style.use('ggplot')
plt.ioff()
```

```python
data_dir = '../data/raw/'
groupby_dir = os.path.join(data_dir, 'groupbycode/all')
col_dtypes = OrderedDict(code=str, date=str, name=str, open=float, high=float, low=float, close=float, volume=int, weekday=int)
```

```python
codes = ['0050', '0051', '0052', '0053', '0054', '0055', '0056', '0057',
         '0058', '0059', '006201', '006203', '006204', '006208', '00690',
         '00692', '00701', '00713']
```

```python
lw = 3
global_props = dict(fontsize=18, kind='box', patch_artist=True, figsize=(10, 5))
boxprops = dict(linewidth=lw)
whiskerprops = dict(linewidth=lw)
capprops = dict(linewidth=lw)
medianprops=dict(linestyle='-', linewidth=lw, color='firebrick')
flierprops=dict(marker='o', markerfacecolor='green', markersize=14, alpha=0.5, linestyle='none')
weekday={1: 'Mon.', 2: 'Tue.', 3: 'Wed.', 4: 'Tur.', 5: 'Fri.', 6: 'Sat.'}

for c in codes:
    etf = pd.read_csv(os.path.join(groupby_dir, f'{c}.csv'),
                      names=col_dtypes.keys(),
                      dtype=col_dtypes, skiprows=1)
#     prev_close = etf.close.shift(periods=1).fillna(etf.open[0])
#     gap = (etf.open - prev_close).rename('gap')
#     gap_weekday = pd.concat([gap, etf.weekday], axis=1)
    diff_weekday = pd.concat([pd.Series(etf.close - etf.open, name='dif'), etf.weekday], axis=1)
    w = []
    for name, df in diff_weekday.groupby('weekday'):
        w.append(df.dif.reset_index(drop=True).rename(f'{weekday[name]}'))
    p = pd.concat(w, axis=1)
    
    fig = ez_plot(p, **global_props,
                  title_prop=dict(label=f'{c}', size=24),
                  legend_prop=None,
                  ylim=[-1, 1],
                  boxprops=boxprops, capprops=capprops,
                  whiskerprops=whiskerprops,
                  medianprops=medianprops,
                  flierprops=flierprops)
    display(HTML(f'<h1>{c}</h1>'))
#     fig.savefig(f'static/weekday_pattern_{c}.jpg')
    display(fig)
    display(p.mean(axis=0).to_frame(name='mean').T)
    display(p.std(axis=0).to_frame(name='std').T)
    display(p.median(axis=0).to_frame(name='median').T)
    display(p.agg(lambda col: (col >= 0.0).sum() / (len(col) - col.isnull().sum())).to_frame(name='不跌的機率').T)
    print('\n' * 2, '=-' * 50, '\n' * 2)
```
